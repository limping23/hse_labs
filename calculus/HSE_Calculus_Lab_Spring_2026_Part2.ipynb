Найти точки экстремума следующей функции: $$u = x^2 + z^2 - xy + x - 2z$$

```python
x = Symbol('x')
y = Symbol('y')
z = Symbol('z')


u_1 = x**2 + z**2 - x*y + x - 2 * z

answer_1 = 0 # В эту переменную вместо нуля присвойте искомые точки экстремума; type(answer_1) = <class 'sympy.sets.sets.FiniteSet'>

x_diff = diff(u_1, x)
y_diff = diff(u_1, y)
z_diff = diff(u_1, z)
stat_points = nonlinsolve([x_diff, y_diff, z_diff], [x, y, z])
delta = Matrix([[u_1.diff(x, 2), u_1.diff(x, y), u_1.diff(x, z)], [u_1.diff(x, y), u_1.diff(y, 2), u_1.diff(y, z)],  [u_1.diff(x, z), u_1.diff(y, z), u_1.diff(z, 2)]]).det()
for i in stat_points:
    x0, y0 = i[0], i[1]
    deltav = delta.subs({x: x0, y: y0})
    if deltav<0:
        print(f"No extremum, point {stat_points} is saddle")
    else:
        answer_1 = (x0, y0)
```

На сетке $[-3, 3] \times [-3, 3]$ найдите локальные минимумы и максимумы с помощью градиентного спуска у следующей функции:

$$
f(x, y) = \cos(0.9(x^2 + y^2)) \cdot \sin(x + y) + 0.05(x^2 + y^2)
$$


```python
# Сетка
x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)

xs = Symbol("x")
ys = Symbol("y")

f = Cos(0.9*(xs**2+ys**2))*Sin(xs+ys)+0.05*(xs**2+ys**2)
dx = simplify(diff(f, xs))
dy = simplify(diff(f, ys))
dx_f = lambdify((xs, ys), dx, modules='numpy')
dy_f = lambdify((xs, ys), dy, modules='numpy')
f_np = lambdify((xs, ys), f, modules='numpy')

def grad_f(x_np, y_np):
    return np.array([dx_f(x_np, y_np), dy_f(x_np, y_np)])

def gradient_descent(start, lr=0.05, eps=1e-5, max_iter=100):
    x = np.array(start)
    trajectory = [x.copy()]
    for _ in range(max_iter):
        grad = grad_f(*x)
        x = x - lr * grad
        trajectory.append(x.copy())
        if np.linalg.norm(grad) < eps:
            break
    return np.array(trajectory)

def gradient_ascent(start, lr=0.05, eps=1e-5, max_iter=100):
    x = np.array(start)
    trajectory = [x.copy()]
    for _ in range(max_iter):
        grad = grad_f(*x)
        x = x + lr * grad
        trajectory.append(x.copy())
        if np.linalg.norm(grad) < eps:
            break
    return np.array(trajectory)

start_points = [
    [0, 0],
    [0.8, 0.8],
    [-1.0, 1.0],
    [1.5, -1.5],
    [-2.0, -1.5],
    [0.0, 2.0],
    [2.5, 2.5],
    [-2.8, 2.8],
    [-1.8, 2.5],
    [0.0, -2.5],
    [2.0, 0.0],
    [1, 1]
]

min_extremums_6 = [] # в эту переменную присвойте минимумы (x, y), каждая точка это кортеж из двух чисел (x, y)
max_extremums_6 = [] # в эту переменную присвойте максимумы (x, y), каждая точка это кортеж из двух чисел (x, y)

for pt in start_points:
    traj = gradient_descent(pt) 
    min_extremums_6.append(tuple(traj[-1]))
    traj1 = gradient_ascent(pt)
    max_extremums_6.append(tuple(traj1[-1]))

print(min_extremums_6)
print(max_extremums_6)
```

```python
def extrem(f, x, y, p):
    dx = diff(f, x)
    dy = diff(f, y)
    stat_points = solve([dx, dy], [x, y], dict=True)
    dxx = diff(f, x, 2)
    dxy = diff(f, x, y)
    dyy = diff(f, y, 2)
    Delta = Matrix([[dxx, dxy], [dxy, dyy]]).det()
    res = []
    for i in stat_points:
        x0, y0 = i.values()
        Deltav = Delta.subs({x: x0, y: y0})
        dxx_v = simplify(dxx.subs({x: x0, y: y0}))
        try:
            delta_num = float(Deltav)
            dxx_num = float(dxx_v)
        except TypeError:
            continue
        if delta_num>0:
            if dxx_num<0:
                if p:
                    for j in p:
                        if x0!=j[0] and y0!=j[1]:
                            res.append((x0, y0, "max"))
                else:
                    res.append((x0, y0, "max"))
            elif dxx_num>0:
                if p:
                    for j in p:
                        if x0!=j[0] and y0!=j[1]:
                            res.append((x0, y0, "min"))
                else:
                    res.append((x0, y0, "min"))
    if res:
        return res
    else:
        return "No extremum"
        

x, y = symbols('x y', real=True)
z1 = x**2 + x*y + y**2 - 2*x - y # x, y - любые
excluded1 = []
print("Z1")
print("x, y - любые")
print(extrem(z1, x, y, excluded1))

x, y = symbols('x y', positive=True) 
z2 = x**3 * y**2 * (6 - x - y) # x, y - любые
excluded2 = []
print("Z2")
print("x, y - любые")
print(extrem(z2, x, y, excluded2))

x, y = symbols("x y", real=True)
z3 = 8/x + x/y + y # x!=0, y!=0
excluded3 = [(0, 0)]
print("Z3")
print("x!=0; y!=0")
print(extrem(z3, x, y, excluded3))

x, y = symbols("x y", real=True)
z3 = (1 + x - y) / Sqrt(1+x**2+y**2)
domainz = (solve(1 + x**2 + y**2 > 0, x)) #x, y - любое
excluded4 = []
print("Z4")
print("x, y - любые")
print(extrem(z3, x, y, excluded4))
```

Что такое нейронная сеть? В общем, *нейронная сеть* — это модель, состоящая из «нейронов» (элементарных блоков), организованных в слои и связанных весами. Каждый нейрон может делать линейное преобразование входа (умножение на вес + смещение), а затем применять нелинейную функцию (ReLU, сигмоида и пр.). В реальных нейронных сетях бывают десятки и сотни слоёв, и тысячи–миллионы нейронов.

в этой задаче мы берём супер упрощённую версию, чтобы увидеть сам принцип: как при помощи градиентного спуска «обучаются» веса сети, минимизируя ошибку.

Рассмотрим следующую простую архитектуру: допустим что у нас есть набор точек $\{x_i, y_i\}_{i=1}^{n}$ и мы хотим научиться предсказывать по $x_i$ переменную $y_i$.
Рассмотрим простую архитектуру, которая является трехслойной нейросетью:
1. На первом слое мы делаем линейное преборазования нашей переменной $x_i:$ $$z_i = w_1 \cdot x + b_1$$
2. Далее применяем нелинейную функцию к $z_i$, например ReLU:
   $$
   h_i = \mathrm{ReLU}(z) = \max(0,z_i).
   $$
   - $\mathrm{ReLU}$ — простейшая нелинейная функция, которая равна $0$ при $z<0$ и $z$ при $z>0$.  
3. Применяем еще один линейный слой к $h_i$, получая итоговое предсказание $y_i'$
   $
   y_i' = w_2 \cdot h_i + b_2.
   $

В данном случе наша модель имеет 4 обучаемых параметра: $(w_1,b_1,w_2,b_2)$ оптимизируясь по которым мы можем предложить модель, которая наилучшим образом предсказывает $y_i$ по известному $x_i$.

Допустим, мы имеем два обучающих примера $(x_1,y_1)$ и $(x_2,y_2)$. Для каждого примера рассчитываем «прогноз» $y'_i$. Допустим, мы хотим уменьшить среднеквадратичную ошибку предсказания и реального значения:
$$
L(w_1,b_1,w_2,b_2)
=
\frac12 \Bigl[(\hat{y}_1 - y_1)^2 + (\hat{y}_2 - y_2)^2\Bigr],
$$
где
$
\hat{y}_i = w_2 \,\max(0,\;w_1\,x_i + b_1) + b_2.
$

Вашей задачей в данном задании будет решить следующую задачу минимизации:
$
\min_{w_1,b_1,w_2,b_2} \;L(w_1, b_1, w_2, b_2).
$
