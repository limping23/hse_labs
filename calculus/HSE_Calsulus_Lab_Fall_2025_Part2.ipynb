```python
x = Symbol('x')

def closest_point(f_1, x_0, y_0):
  result = () #type(result) = tuple = (x-координата искомой точки, у-координата искомой точки). В эту переменную запиши координаты искомой точки.
  dist = (x-x_0)**2+(f_1-y_0)**2
  d_dist = diff(dist, x)
  x1 = solve(d_dist, x)[0]
  result = (x1, f_1.subs(x, x1))
  return result
```

Вам дана функция `f_2`. Найдите уравнение касательной к `f_2` в точке `x_2`.

```python
x = Symbol('x')

def GetTangent(g_2, x_2):
  result = 0 #type(result) = sympy.core.add.Add. В эту переменную запишите искомое уравнение касательной (вида например: result = 5 + 10*x)
  y_2 = g_2.subs(x, x_2)
  k = diff(g_2, x).subs(x, x_2)
  result = y_2+k*(x-x_2)
  return result
```

```python
#TEST CASE

try:
  assert(GetTangent(x**2, -5) == -25 - 10*x)
  print("GOOD")
except:
  print("BAD")
```
